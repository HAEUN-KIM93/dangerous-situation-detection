```python

import cv2
import os
import numpy as np

   
```

```python
class object_detection:
    def __init__(self):
        self.model=YOLO("dangerous.pt")
    def predict(self,filePath,receiver):
       
        class_names = model.names 
        
        print(filePath)

        if os.path.isfile(filePath):	
            
            cap = cv2.VideoCapture(filePath)
        else:
            print("file does not exist.")  

        
        frameWidth = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))	
        frameHeight = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))	
        
        frame_size = (frameWidth, frameHeight)
        


        detected_class=np.array([],dtype=int)
        while True:
            
            retval, frame = cap.read()
            if not(retval):	
                break  
            result=model(source=frame ,show=True , conf=0.5,save=True)
            annotated_frame=result[0].plot()
            f_result=(result[0].boxes.cpu().numpy())
            for i in range(len(f_result.cls)):
                detected_class=np.append(detected_class,class_names[int(f_result.cls[i])])
                detected_class=np.unique(detected_class)
             
            cv2.imshow('annotated_video', annotated_frame)	
            key = cv2.waitKey(1)
            
          
            if key == 27:
                break	
            
            
        cv2.destroyAllWindows()
        print(detected_class)
        
        if len(detected_class)!=0: 
            sending=mail('alert')
            sending.sending_email(detected_class,receiver)
            


        

```
